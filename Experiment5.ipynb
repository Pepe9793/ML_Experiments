````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
df = pd.read_csv("diabetes.csv")
df.head()
X = df.drop("Outcome", axis=1)
y = df["Outcome"]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
svm = SVC(kernel="rbf", gamma="scale", C=1)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap="Greens", fmt='d')
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
print(classification_report(y_test, y_pred))

````

## Experiment Result Report

The Support Vector Machine (SVM) model was trained and evaluated on the Diabetes dataset. The goal was to predict the `Outcome` (diabetes presence) based on the provided features.

### Model Performance Summary

*   **Overall Accuracy**: The model achieved an accuracy of **72.73%** on the test set. This means that approximately 73 out of 100 predictions made by the model were correct.

### Confusion Matrix

The confusion matrix provides a detailed breakdown of correct and incorrect predictions:

```
[[81, 18]
 [24, 31]]
```

*   **True Negatives (TN)**: 81 (The model correctly predicted 81 instances as non-diabetic).
*   **False Positives (FP)**: 18 (The model incorrectly predicted 18 instances as diabetic when they were non-diabetic - Type I error).
*   **False Negatives (FN)**: 24 (The model incorrectly predicted 24 instances as non-diabetic when they were diabetic - Type II error).
*   **True Positives (TP)**: 31 (The model correctly predicted 31 instances as diabetic).

### Classification Report

The classification report provides precision, recall, and f1-score for each class:

```
              precision    recall  f1-score   support

           0       0.77      0.82      0.79        99
           1       0.63      0.56      0.60        55

    accuracy                           0.73       154
   macro avg       0.70      0.69      0.70       154
weighted avg       0.72      0.73      0.72       154
```

**Class 0 (Non-Diabetic)**:
*   **Precision**: 0.77 - When the model predicts non-diabetic, it is correct 77% of the time.
*   **Recall**: 0.82 - The model correctly identifies 82% of all actual non-diabetic cases.
*   **F1-Score**: 0.79 - The harmonic mean of precision and recall for non-diabetic predictions.

**Class 1 (Diabetic)**:
*   **Precision**: 0.63 - When the model predicts diabetic, it is correct 63% of the time.
*   **Recall**: 0.56 - The model correctly identifies 56% of all actual diabetic cases.
*   **F1-Score**: 0.60 - The harmonic mean of precision and recall for diabetic predictions.
